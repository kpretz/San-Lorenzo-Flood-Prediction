## Random Forest Model
Build Random forest model with several variables lagged, to get general idea of feature importance and model fit capabilities as a baseline. 

```python
data_rfr = data.copy()
months_to_lag = [1, 2, 3, 4, 6] 
drought_index_ints = [0, 1, 2, 3, 4]

for m in months_to_lag:
    data_rfr['lag_precip_' + str(m) + 'months'] = data_rfr['precip_in_inc'].shift(-30*96*m) # shift data back, 30 days in a month
    for d in drought_index_ints:
        data_rfr['lag_d' + str(d) + '_' + str(m) + 'months'] = data_rfr['D' + str(d)].shift(-30*96*m) #shift back data

days_to_lag = [1, 2, 5, 14, 7]
for h in days_to_lag:
    data_rfr['lag_precip_' + str(h) + 'days'] = data_rfr['precip_in_inc'].shift(h*96) # shift data back h hours


data_rfr.dropna(inplace=True) #drop null values created with shift functions - will be most recent 4 months
data_rfr.head()
# data_rfr = data_rfr.loc[data_rfr['flood_stage_SC'] >= 1, :]
```

```python
def run_rfr_cv(data, target, grid, strat):
    X = data.drop(['stage_ft_BT', 'stage_ft_SC', 'flood_stage_SC', 'flood_stage_BT', 'q_cfs_peak_SC_int', 
                   'q_cfs_peak_BT_int', 'cy', 'wy', 'month'], axis=1)
    y = data[target]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=777, stratify=strat)
    
    
    rf_cv = GridSearchCV(estimator=RandomForestRegressor(), 
                         param_grid=grid, cv=3)
    rf_cv.fit(X_train, y_train)
    rfr_best = rf_cv.best_estimator_
    y_pred = rfr_best.predict(X_test)
    
    print('training model score: ', rfr_best.score(X_train, y_train))
    print('test model score: ', rfr_best.score(X_test, y_test))
    
    #next, we'll take a look at feature importance
    importances = rfr_best.feature_importances_
    forest_importances = pd.Series(importances, index=X_train.columns)
    
    fig, ax = plt.subplots()
    forest_importances.sort_values().plot.bar(ax=ax)
    
    ## plot results
    f, ax1 = plt.subplots(1, 1, figsize=(6, 5))
    
    ax1.scatter(y_pred, y_test)
    # ax1.plot((0, 20000), (0, 20000), 'k', linestyle='--')
    ax1.set_xlabel('predicted')
    ax1.set_ylabel('observed')
    ax1.set_title(target)
    ax1.set_aspect('equal', adjustable='box')
    ax1.grid(visible=True)
    lims = [
        np.min([ax1.get_xlim(), ax1.get_ylim()]),  # min of both axes
        np.max([ax1.get_xlim(), ax1.get_ylim()]),  # max of both axes
    ]
    ax1.plot(lims, lims, 'k-', alpha=0.75, zorder=0, linestyle='--')
    
    print(rf_cv.best_params_)
    print('MSE: ', mean_squared_error(y_test, y_pred))
    return y_pred, rf_cv.cv_results_
```

```python
data_daily = data_rfr.resample('1d').max()
grid = {
        'max_depth': [5, 6, 8, 2, 3, 4],
        'n_estimators': [100, 30, 50, 20, 10], 
        'random_state': [777]
    }

y_pred_SC, results_SC = run_rfr_cv(data_daily, 'q_cfs_peak_BT_int', grid, data_daily['flood_stage_BT'])
```

## Neural Net Model

```python
target = 'stage_ft_SC'
data_nn = data_daily.copy()
data_nn.head()
```

```python
# split test/train data
s_scaler = MinMaxScaler()
s_scaler.set_output(transform='pandas')

p_scaler = PowerTransformer()
s_scaler.set_output(transform='pandas')

scaler_pipe = Pipeline(steps=[('scale', s_scaler),
                              ('power', p_scaler)])
scaler_pipe.set_output(transform='pandas')



# split test/train data
X = data_nn.drop(['stage_ft_BT', 'stage_ft_SC', 'flood_stage_SC', 'flood_stage_BT', 'q_cfs_peak_SC_int', 
                   'q_cfs_peak_BT_int', 'cy', 'wy', 'month'], axis=1)
y = data_nn[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=777, stratify=data_nn['flood_stage_SC'])

X_train = scaler_pipe.fit_transform(X_train)
X_test = scaler_pipe.fit_transform(X_test)

# y_train = scaler_pipe.fit_transform(np.ravel(y_train))
# y_test = scaler_pipe.transform(np.array(y_test))

model = Sequential()
n_col = len(X_train.columns)

model.add(Input(shape=(n_col,)))
model.add(Dense(1000, activation='linear', input_shape=(n_col,)))
model.add(Dense(500, activation='relu', input_shape=(n_col,)))
model.add(Dense(500, activation='relu', input_shape=(n_col,)))
model.add(Dense(200, activation='relu', input_shape=(n_col,)))
model.add(Dense(200, activation='relu', input_shape=(n_col,)))

model.add(Dense(1))
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=200, verbose=False)
```

```python
y_pred_test = model.predict(X_test)
# y_pred_test = scaler_pipe.inverse_transform(y_pred_test)

f, ax = plt.subplots(1, 1)
ax.scatter(y_pred_test, y_test)
# plt.plot([0, 250000], [0, 250000], linestyle='--')
ax.set_ylabel('observed')
ax.set_xlabel('predicted')
ax.set_aspect('equal', adjustable='box')
ax.grid(visible=True)
lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes
]
ax.plot(lims, lims, 'k-', alpha=0.75, linestyle='--')
print('MSE: ', mean_squared_error(y_test, y_pred_test))
```

## Random Forest Classifier of flood stage



```python
def rfc_model(data, target, grid, strat):
    X = data.drop(['stage_ft_BT', 'stage_ft_SC', 'flood_stage_SC', 'flood_stage_BT', 'q_cfs_peak_SC_int', 'cy', 'q_cfs_peak_BT_int', 
                   'month', 'wy'], axis=1)
    y = data[target]
    le = LabelEncoder()
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=777, stratify= strat)
    
    
    rfc_cv = GridSearchCV(estimator=RandomForestClassifier(), 
                             param_grid=grid, cv=3)
    rfc_cv.fit(X_train, y_train)
    rfc_best = rfc_cv.best_estimator_
    y_pred = rfc_best.predict(X_test)
    
    print('training model score: ', rfc_best.score(X_train, y_train))
    print('test model score: ', rfc_best.score(X_test, y_test))
    
    #next, we'll take a look at feature importance
    importances = rfc_best.feature_importances_
    forest_importances = pd.Series(importances, index=X_train.columns)
    
    fig, ax = plt.subplots()
    forest_importances.sort_values().plot.bar(ax=ax)
        
    disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred))
    
    disp.plot()
    print(rfc_cv.best_params_)
    print('accuracy = ', accuracy_score(y_test, y_pred))
    print('precision = ', precision_score(y_test, y_pred, average='weighted'))
    print('recall = ', recall_score(y_test, y_pred, average='weighted'))
    print('MSE: ', mean_squared_error(y_test, y_pred))
    return y_pred, rfc_cv
```

```python
data_rfc = data_rfr.loc[data_rfr['flood_stage_BT'] >= 1, :]

grid = {
            'max_depth': [5, 6, 8, 4, 10, 12, 13, 15],
            'n_estimators': [50, 30, 100, 200, 300, 500, 750], 
            'random_state': [777]
        }

y_pred_SC, rfc_cv = rfc_model(data_rfc, 'flood_stage_BT', grid, data_rfc['flood_stage_BT'])
```

```python
data_rfc = data_rfr.loc[data_rfr['flood_stage_SC'] >= 1, :]

grid = {
            'max_depth': [5, 6, 8, 7, 4, 10, 12, 15],
            'n_estimators': [50, 30, 100, 200, 300, 350, 400, 500], 
            'random_state': [777]
        }

y_pred_SC, rfc_cv = rfc_model(data_rfc, 'flood_stage_SC', grid, data_rfc['flood_stage_SC'])
```

## Logistic Regression

```python
def run_lr_model(data, target, grid, strat):
    
    X = data.drop(['stage_ft_BT', 'stage_ft_SC', 'flood_stage_SC', 'flood_stage_BT', 'q_cfs_peak_SC_int', 'cy', 'q_cfs_peak_BT_int'], axis=1)
    y = data[target]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=777, stratify=strat)
    
    scaler = StandardScaler()
    scaler.set_output(transform='pandas')
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.fit_transform(X_test)
    
    # define model
    
    lr_cv = GridSearchCV(estimator=LogisticRegression(), 
                             param_grid=grid, cv=3)
    lr_cv.fit(X_train, y_train)
    lr_best = lr_cv.best_estimator_
    y_pred = lr_best.predict(X_test)
        
    print('training model score: ', lr_best.score(X_train, y_train))
    print('test model score: ', lr_best.score(X_test, y_test))
    
    
    disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred))
    
    disp.plot()
    print(lr_cv.best_params_)

    print('accuracy = ', accuracy_score(y_test, y_pred))
    print('precision = ', precision_score(y_test, y_pred, average='weighted'))
    print('recall = ', recall_score(y_test, y_pred, average='weighted'))
    print('MSE: ', mean_squared_error(y_test, y_pred))

    return y_pred, lr_cv
```

```python
data_onlyfloods = data_rfr.loc[data_rfr['flood_stage_BT'] >= 1, :]
grid = {
            'C': [0.5, 1, 2, 5, 6, 7, 8, 10],
            'class_weight': ['balanced', None], 
            'max_iter': [1000, 5000, 600, 10000, 100000],
            'random_state': [777]
        }
y_pred, lr_cv = run_lr_model(data_onlyfloods, 'flood_stage_BT', grid, data_onlyfloods['flood_stage_BT'])
```
